```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv("/content/heart.csv")

# feature and target split.
X = heart_data.drop(columns="target", axis=1)
Y = heart_data["target"]

# Standardizing the dataset.
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
# instantiate the logistic_regression model.
model = svm.SVC(kernel="linear")
# fit the dataset
model.fit(X_train, Y_train)
```

```python
# Accuracy on training data.
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)

# Accuracy on test data.
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(testing_data_accuracy)

# Accuracy is not good due to small dataset.
# test_accuracy ~= train_accuracy, 
# This means that we have avoided overtraining and undertraining.
```

```python
input_data = (62, 0, 0, 140, 268, 0, 0, 160, 0, 3.6, 0, 2, 2)
# change the input to a numpy array.
input_data_as_np_array = np.asarray(input_data)
# reshape the np array to predict over one datapoint only.
input_data_reshaped = input_data_as_np_array.reshape(1, -1)

# Prediciton
prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0]:
    print("The person has a heart disease")
else :
  print("The person has no heart disease")
```
