```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
diabetes_dataset = pd.read_csv("/content/diabetes.csv")

# Printing the first 5 rows of our dataframe for preview.
diabetes_dataset.head()
```

```python
# Separation of data and labels.
X = diabetes_dataset.drop(columns="Outcome", axis=1)
Y = diabetes_dataset["Outcome"]
X.head()
```

```python
# Instantiate scaler and fit dataset(X).
scaler = StandardScaler()
standardized_data = scaler.fit_transform(X)
print(standardized_data)

# Assign X to the standardized data.
X = standardized_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, 
                                              stratify=Y, random_state=2)
print("Shape for X dataset: ", X.shape, X_train.shape, X_test.shape)
print("Shape for Y dataset: ", Y.shape, Y_train.shape, Y_test.shape)
```

```python
# Instatiate the classifier
classifier = LogisticRegression()

# Training the Logistic Regression classifier
classifier.fit(X_train, Y_train)
```

```python
# Accuracy score on train-dataset
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Initial Accuracy on train_set: ", training_data_accuracy)
```

```python
# Accuracy score on test-dataset
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Initial Accuracy on test_set: ", testing_data_accuracy)
```

```python
input_data = [5, 166, 72, 19, 175, 25.8, 0.587, 51]

# changing input data to np.array
input_data_as_np_array = np.asarray(input_data)

# reshape the array for prediction of one instance only.
input_data_reshaped = input_data_as_np_array.reshape(1, -1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

# prediction 
prediction = classifier.predict(std_data)
print(prediction)
# we are using prediction[0] because function returns prediction
# as a list with only one element in it.
if prediction[0]:
  print("Patient has diabetes")
else:
  print("Patient doesn't have diabetes")
```
